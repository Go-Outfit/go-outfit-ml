```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import os
from tensorflow import keras
```

```python
model = keras.models.load_model('detect_image_user.model')
```

```python
def predict_image(file_path, model):
    img = tf.keras.utils.load_img(file_path, target_size=(150, 150))
    x = tf.keras.utils.img_to_array(img)
    x/=255
    x = np.expand_dims(x, axis = 0)
    
    images = np.vstack([x])
    classes = model.predict(images, batch_size = 100)
    print(classes)
    plt.imshow(tf.keras.utils.load_img(file_path,False,color_mode="rgba"))
    plt.show()
    labels = ["celana","dress","heels","jaket","jas","kaos","rok","sandal","sepatu"]
    output_class=labels[np.argmax(classes)]
    if output_class == "dress" and "jaket" and "kaos" and "jas":
        print("This is {} from atasan".format(output_class))
    elif output_class == "celana" and "rok":
        print("This is {} from bawahan".format(output_class))
    elif output_class == "sandal" and "heels" and "sepatu":
        print("This is {} from alas kaki".format(output_class))
    else:
        print("Please reupload your picture!") 
    
```

```python
predict_image("./6adbdd7bd6751ed01eed2d1eddf343d7.jpg",model)
```
